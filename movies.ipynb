```python
import pandas as pd
```

```python
movies=pd.read_csv("F:\\food_geoloc\\tmdb_5000_movies.csv")
credits=pd.read_csv("F:\\food_geoloc\\tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
credits_col_renamed=credits.rename(index=str,columns={"movie_id":"id"})
movies_merge=movies.merge(credits_col_renamed,on='id')
movies_merge.head()
```

```python
movies_clean=movies_merge.drop(columns=["homepage","title_x","title_y","status","production_countries"])
movies_clean.head()
```

```python
movies_clean.info() #to check 'NAN' or null values,here no null values
```

```python
v=movies_clean['vote_count']                #no. of votes
R=movies_clean['vote_average']              #avg of no. of votes
C=movies_clean['vote_average'].mean()       #mean vote of movies
m=movies_clean['vote_count'].quantile(0.70) #consider movies having more tha 70 percentile
```

```python
movies_clean['weighted_average']=((R*v)+(C*m))/(v+m) #creating weightes_avg column in table
```

```python
movies_clean.head()
```

```python
movies_sorted=movies_clean.sort_values('weighted_average',ascending=False)#decreasing order
movies_sorted[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)#recommendation acc to weighted avg or user-based recommendation or based on no. of people voted the movie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movies_sorted.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best movies by average votes',weight='bold')
plt.xlabel('Weighted Average Score',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.savefig('best.movies.png')
```

```python
popularity=movies_sorted.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10),data=popularity)

plt.title('Most popular by votes',weight='bold')
plt.xlabel('score of popularity',weight='bold')
plt.ylabel('Movies Title',weight='bold')
plt.savefig('best_popular_movies.png')
```

# Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler 

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_clean[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_clean[['normalized_weighted_average','normalized_popularity']]=movie_normalized_df
```

```python
movies_clean.head()
```

```python
movies_clean['score']=movies_clean['normalized_weighted_average']*0.5 + movies_clean['normalized_popularity']*0.5
movies_clean=movies_clean.sort_values(['score'],ascending=False)
movies_clean[['original_title','normalized_weighted_average','normalized_popularity','score']].head(20)
```

```python
scored_df=movies_clean.sort_values('score',ascending=False)
plt.figure(figsize=(16,6))
ax=sns.barplot(x=scored_df['score'].head(10),y=scored_df['original_title'].head(10),data=scored_df)

#plt.xlim(3.55,5.25)
plt.title('Best rated and most popular blend',weight='bold')
plt.xlabel('Score',weight='bold')
plt.ylabel('Movie Title',weight='bold')

plt.savefig('scored_movies.png')
```
